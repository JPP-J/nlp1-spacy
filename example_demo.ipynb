## Example 2
To check mode in NLP task 
- `NER`  Detect named entities (e.g., names, places, dates).
- `token`  Split text into words/tokens.
- `POS`  Tag words with parts of speech (noun, verb, etc.).
- `DEP`  Show grammatical structure (subject, object, etc.).
- `LEMMA`  Reduce words to base form (e.g., "running"  "run").
- `MORPH`  Show word features (tense, number, gender, etc.).
- `SENT`  Split text into sentences.
- `SIM`  Compute similarity between texts.
- `SENTIMENT`}  Analyze sentiment (polarity, subjectivity).

## Example3
To check pipeline 
- `tok2vec`  Word vector representations (used for context in downstream tasks).
- `tagger`  Part-of-speech tagging (noun, verb, etc.).
- `parser`  Dependency parsing (grammatical structure).
- `attribute_ruler`  Adds or overrides token attributes (custom rules).
- `lemmatizer`  Converts words to their base forms (e.g., "running"  "run").
- `ner`  Named Entity Recognition (detects names, locations, etc.).- 
- `spacytextblob`  Sentiment analysis (polarity & subjectivity).
